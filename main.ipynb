## tools

````python
def print_art(art):
    d_set, d_slot, d_main, d_level = \
        art["setKey"], art["slotKey"], art["mainStatKey"], art["level"]
    print(d_set, d_slot, d_level, d_main,
          ", ".join([f"{x['key']}: {x['value']}" for x in art["substats"]]))


print_bak = print


def print_off():
    global print

    def do_nothing(*args, **kargs):
        pass
    print = do_nothing


def print_on():
    global print
    print = print_bak
````

## Rule handler

````python
w4_list = [0.4, 0.3, 0.2, 0.1]
w4_list = [0.25 for _ in range(4)]
w4_list = [0.256, 0.252, 0.248, 0.244]


w10_list = [1] * 10  # 正常情况就是用6或7个
w10_list[:4] = 1.2, 1.1, 1, 1  # 正常情况就是用6或7个


def match(r_sub_map, art_sub):
    cur_score = 0
    k_socre_list = []
    for kv in art_sub:
        k, v = kv["key"], kv["value"]
        k_score = r_sub_map.get(k, None)
        if k_score is not None:
            cur_score += v / single[k] * base_score * k_score
            k_socre_list.append(k_score)
        else:
            k_socre_list.append(0)

    return cur_score, k_socre_list


def stat4score(r_sub_map, art_sub_map, art_main):
    left_keys = sub_option - set([x["key"] for x in art_sub_map])
    left_keys.remove(art_main)
    left_k_score = [r_sub_map.get(k, 0) for k in left_keys]
    left_k_score = sorted(left_k_score, reverse=True)
    print(f"{left_k_score=}")
    score_sub_4 = 0
    w_count = 0
    for i in range(len(left_k_score)):
        score_sub_4 += left_k_score[i] * w10_list[i]
        w_count += w10_list[i]
    return score_sub_4 / w_count


def rule_handler(rule, art):

    r_set, r_slot, r_main, r_sub = rule
    d_set, d_slot, d_main, d_sub = \
        art["setKey"], art["slotKey"], art["mainStatKey"], art["substats"]

    if (not r_set or d_set in r_set) and \
       (not r_slot or d_slot in r_slot) and \
       (not r_main or d_main in r_main):
        print("match")
        print(rule)
        print_art(art)
        cur_score, k_socre_list = match(r_sub, d_sub)
        print(f"{cur_score=} ")
        left_time = 5 - art["level"] // 4
        if len(d_sub) == 3:
            stat4 = stat4score(r_sub, d_sub, d_main)
            k_socre_list.append(stat4)
            left_time -= 1
            cur_score += stat4 * base_score
            print(f"{stat4=} ")
        k_socre_list = sorted(k_socre_list, reverse=True)
        print(f"{k_socre_list=} ")
        weighed_score = 0
        for i in range(4):
            weighed_score += k_socre_list[i] * w4_list[i]
        total_score = cur_score + weighed_score * base_score * left_time
        print(f"{total_score=} ")
        return total_score
    return 0


def handle_rule_arts(rule, arts):
    res = []
    for idx in range(len(arts)):
        score = rule_handler(rule, arts[idx])
        if score > 0:
            res.append((idx, score))
    return sorted(res, key=lambda x: x[1], reverse=True)
````

## test

````python
test_rules = [
    [['追忆'], ['花', '羽'], None, {'暴击': 1.0, '爆伤': 1.0, '大攻击': 0.6, '精通': 0.8}],
    [[乐团, 饰金, 绽放], ['花', '羽'], None, {'精通': 1.0, '充能': 1.0}]
]


def test(test_rules):
    print(f"{w4_list=}")
    print(f"{w10_list=}")
    for r in test_rules:
        print(r)
        print_off()
        res = handle_rule_arts(r, artifacts)
        print_on()
        count = 0
        for idx, score in res[:]:
            if score <= 15:
                break
            print(count, round(score, 1), end=" ")
            print_art(artifacts[idx])

            count += 1


def test2(test_rules):
    print(f"{w4_list=}")
    print(f"{w10_list=}")
    for r in test_rules:
        print(r)
        handle_rule_arts(r, artifacts)


with open("../ys-git/log", "w", encoding="utf-8") as stdout_f:
    old_stdout = sys.stdout
    sys.stdout = stdout_f
    test2(test_rules)
    sys.stdout = old_stdout
````
